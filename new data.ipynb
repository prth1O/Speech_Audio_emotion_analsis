```python
#change seseon name
cat="E:/Work/Audio DATA/IEMOCAP_full_release_withoutVideos/IEMOCAP_full_release/Session5/dialog/EmoEvaluation/Categorical/"
categories=glob.glob(cat+'*.txt')
```

```python
data={'name':[],
     'Emot':[]}
```

```python

pat=re.compile(r'\b\w+(?=;)')
pl=re.compile(r'\w+(?=\s:)')
for path in categories:
    df=pd.read_csv(path,names=['o'])
    for dat in df['o'].tolist():
        emot=re.findall(pat,dat)
        name=re.findall(pl,dat)
        if len(emot) >= 1:
            for i in emot:
                data['name'].extend(name)
                data['Emot'].append(i)
    
```
